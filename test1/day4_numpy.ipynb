```python
#水平连接
#axis=0行
#axis=1列
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
c = np.hstack([a,b])
print(c)
c = np.concatenate([a,b],axis=1)
print(c)
```

```python
#垂直
c = np.vstack([a,b])
print(c)
c = np.concatenate([a,b],axis=0)
print(c)
```
